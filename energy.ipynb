```python
x=df.drop('price_actual',axis=1)
y=df['price_actual']
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Define the number of bins or categories
num_bins = 5

# Use numpy's digitize function to bin the continuous values into discrete categories
y_discrete = np.digitize(y, bins=np.linspace(y.min(), y.max(), num_bins + 1))
x_train, x_test, y_train, y_test = train_test_split(x, y_discrete, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import  StandardScaler
from sklearn.metrics import accuracy_score
```

```python
feature=x.columns
feature
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")
```
